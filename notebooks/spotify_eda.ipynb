```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme(style="whitegrid")

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("../data/spotify_tracks.csv")

print("Dataset loaded successfully!")
print("Shape:", df.shape)
```

## Data Cleaning
Removing unnecessary columns, handling missing values, and preparing the dataset for analysis.

```python
# Remove unnecessary column
df.drop(columns=['Unnamed: 0'], inplace=True)

# Remove rows with missing values
df.dropna(inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

print("New Shape:", df.shape)
```

```python
df.isnull().sum().sum()
```

## Dataset Overview
Understanding the structure of the Spotify dataset before visualization.

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names:")
print(df.columns.tolist())
```

## Popularity Distribution
Analyzing how song popularity is distributed across Spotify tracks.

```python
plt.figure(figsize=(10,6))

sns.histplot(df['popularity'], bins=30, kde=True)

plt.title("Distribution of Spotify Song Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Number of Songs")

plt.savefig("../images/popularity_distribution.png",
            bbox_inches="tight")

plt.show()
```

## Genre Analysis

Analyzing the most common music genres in the Spotify dataset and understanding their representation.

```python
top_genres = df['track_genre'].value_counts().head(15)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_genres.values,
    y=top_genres.index
)

plt.title("Top 15 Most Common Spotify Genres")
plt.xlabel("Number of Songs")
plt.ylabel("Genre")

plt.savefig("../images/top_genres.png",
            bbox_inches="tight")

plt.show()
```

## Explicit Content Analysis

Comparing the distribution of explicit and non-explicit tracks.

```python
plt.figure(figsize=(6,6))

df['explicit'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel("")
plt.title("Explicit vs Non-Explicit Songs")

plt.savefig("../images/explicit_content.png",
            bbox_inches="tight")

plt.show()
```

## Correlation Analysis

Examining relationships between Spotify audio features to identify patterns influencing song characteristics.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm',
    annot=False
)

plt.title("Correlation Heatmap of Spotify Features")

plt.savefig("../images/correlation_heatmap.png",
            bbox_inches="tight")

plt.show()
```

## Popularity vs Danceability

Investigating whether more danceable songs tend to be more popular.

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df.sample(5000),
    x='danceability',
    y='popularity',
    alpha=0.5
)

plt.title("Popularity vs Danceability")
plt.xlabel("Danceability")
plt.ylabel("Popularity")

plt.savefig("../images/popularity_vs_danceability.png",
            bbox_inches="tight")

plt.show()
```

## Popularity vs Energy

Identifying artists with the highest number of tracks in the dataset.

## Song Duration Analysis

Analyzing the distribution of track durations.

```python
df['duration_min'] = df['duration_ms'] / 60000

plt.figure(figsize=(10,6))

sns.histplot(
    df['duration_min'],
    bins=40,
    kde=True
)

plt.title("Distribution of Song Duration")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Songs")

plt.xlim(0,10)

plt.savefig("../images/energy_distribution.png",
            bbox_inches="tight")

plt.show()
```

# Key Business Insights

Based on the exploratory analysis of Spotify tracks, several important patterns were identified regarding song popularity, genre distribution, and audio characteristics.

### Insight 1: Popular Songs Are Rare

The popularity distribution indicates that only a small proportion of tracks achieve very high popularity scores, while most songs remain in the low-to-medium popularity range.

### Insight 2: Genre Diversity Is Significant

The dataset contains a wide variety of genres, demonstrating Spotify's diverse music ecosystem and broad listener preferences.

### Insight 3: Danceability Shows Moderate Influence

Songs with higher danceability tend to appear more frequently among popular tracks, suggesting that listener engagement may be linked to dance-friendly music.

### Insight 4: Energy Alone Does Not Guarantee Popularity

While energetic songs are common, the relationship between energy and popularity is not strong enough to conclude that higher energy directly leads to greater success.

### Insight 5: Most Songs Fall Within a Similar Duration Range

The majority of tracks are clustered around standard song lengths, indicating industry consistency in music production.

# Conclusion

This project explored over 114,000 Spotify tracks using exploratory data analysis techniques.

The analysis included data cleaning, statistical summaries, genre analysis, popularity analysis, audio feature exploration, and correlation analysis.

Key findings suggest that song popularity is influenced by multiple factors rather than a single audio characteristic. Danceability, genre diversity, and listener preferences appear to contribute to track success, while most songs follow common duration and production patterns.

This project demonstrates how exploratory data analysis can be used to uncover meaningful insights from large-scale music datasets and support data-driven decision-making.
